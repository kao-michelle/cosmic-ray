```python
from matplotlib import pyplot as plt
import numpy as np
import math
from astropy import units as u
from astropy.constants import c, m_e, sigma_T, alpha

### E_e = (1) 0.1 GeV, (2) 1 GeV, (3) 10 GeV

fig, ax1 = plt.subplots(figsize=[7,5]) 
n_p = np.linspace(10**-3, 10**3, 10**6)

# coefficient of n_p
K_rad1 = (((4/3) * sigma_T / (m_e**2 * c**3) * 0.1*u.GeV * 1*u.eV/u.cm**3)**-1).to(u.Myr) # already in Myr
K_ff1 = (alpha * c * sigma_T).to(u.cm**3/u.Myr)
K_c1 = (30 / (0.1*u.GeV) * m_e * c**3 * sigma_T).to(u.cm**3/u.Myr)

# energy loss timescale
tau_rad1 = K_rad1.value * (8.7*10**-8 * (n_p/(10**5))**(-1/2) * n_p)**0
tau_ff1 = (K_ff1.value * 8.7*10**-8 * (n_p/(10**5))**(-1/2) * n_p)**-1
tau_c1 = (K_c1.value * 8.7*10**-8 * (1/(10**5))**(-1/2) * n_p**(1/2))**-1

# plotting tau vs. n_p
ax1.loglog(n_p, tau_rad1, color = 'dodgerblue', linestyle = '--', label = 'non-thermal emission')
ax1.loglog(n_p, tau_ff1, color = 'dodgerblue', linestyle = '-.', label = 'free-free emission') 
ax1.loglog(n_p, tau_c1, color = 'dodgerblue',linestyle = ':', label = 'coulomb scattering')

ax1.set_title('$E_e$ = 0.1 GeV')
ax1.set_ylabel(r'$\tau$ / Myr', fontsize = 14)
ax1.set_xlabel('$n_p$ / cm$^{-3}$', fontsize = 14)

# plotting thermalization efficiency
f_c = tau_c1**-1 / (tau_c1**-1 + tau_rad1**-1 + tau_ff1**-1)

ax2 = ax1.twinx() # create another y-axis sharing a common x-axis
ax2.plot(n_p, f_c, 'r')
ax2.set_ylabel('$f_C$', color = 'red', fontsize = 14)
ax2.tick_params(axis = 'y', labelcolor = 'red')
ax2.yaxis.set_ticks(np.arange(0, 1.2, 0.2))

fig.subplots_adjust(top = 0.82)
fig.legend()
#plt.savefig('0.1 GeV.pdf')
```

```python
from matplotlib import pyplot as plt
import numpy as np
import math
from astropy import units as u
from astropy.constants import c, m_e, sigma_T, alpha
```

```python
# magnetic field - density relation 
def B(n_p):
    
    n1 = 300 #cm^-3
    B_ICM = (10 * u.uG).to(u.G)
    q = 2/3
    
    if n_p <= n1:
        return B_ICM.value
    else:
        return B_ICM.value * (n_p / n1)**q
```

```python
### energy loss timescale

def t_rad(n_p, E_e):
    U_rad = 0.26*u.eV*u.cm**-3
    K_rad = (((4/3) * sigma_T / (m_e**2 * c**3) * E_e * U_rad)**-1).to(u.Myr) # already in Myr
    return K_rad.value * (8.7*10**-8 * 10**(1/2) * (n_p/(10**5))**(-1/2) * n_p)**0

def t_sync(n_p, E_e):
    U_mag = ((B(n_p)**2 / (8 * math.pi))*u.erg*u.cm**-3).to(u.eV*u.cm**-3)
    K_sync = (((4/3) * sigma_T / (m_e**2 * c**3) * E_e * U_mag)**-1).to(u.Myr) # already in Myr
    return K_sync.value * (8.7*10**-8 * 10**(1/2) * (n_p/(10**5))**(-1/2) * n_p)**0
    
def t_ff(n_p, E_e):
    K_ff = (alpha * c * sigma_T).to(u.cm**3/u.Myr)
    return (K_ff.value * 8.7*10**-8 * 10**(1/2) * (n_p/(10**5))**(-1/2) * n_p)**-1
    
def t_c(n_p, E_e):
    K_c = ((30 / E_e) * m_e * c**3 * sigma_T).to(u.cm**3/u.Myr)
    return (K_c.value * 8.7*10**-8 * 10**(1/2) * (1/(10**5))**(-1/2) * n_p**(1/2))**-1
```

```python
n_p_array = np.logspace(1, 5, 10**4)
E_e = 0.01*u.GeV

tau_rad = np.zeros_like(n_p_array)
tau_sync = np.zeros_like(n_p_array)
tau_ff = np.zeros_like(n_p_array)
tau_c = np.zeros_like(n_p_array)
f_c = np.zeros_like(n_p_array)

for i in np.arange(len(n_p_array)):
    n_p = n_p_array[i]
    
    # energy loss timescale
    tau_rad[i] = t_rad(n_p, E_e)
    tau_sync[i] = t_sync(n_p, E_e)
    tau_ff[i] = t_ff(n_p, E_e)
    tau_c[i] = t_c(n_p, E_e)
    
    #print(tau_rad[i], tau_sync[i], tau_ff[i], tau_c[i])
    
    # heating efficiency
    f_c[i] = tau_c[i]**-1 / (tau_c[i]**-1 + tau_rad[i]**-1 + tau_sync[i]**-1 + tau_ff[i]**-1)
    #print(f_c[i])
    #print()
    
```

```python
fig, ax1 = plt.subplots(figsize=[8,5.5]) 

# plotting tau vs. n_p
ax1.loglog(n_p_array, tau_rad, color = 'dodgerblue', linestyle = '--', label = 'radiative inverse-Compton')
ax1.loglog(n_p_array, tau_sync, color = 'mediumslateblue', linestyle = '--', label = 'radiative synchrotron')
ax1.loglog(n_p_array, tau_ff, color = 'mediumorchid', linestyle = '--', label = 'freefree bremsstrahlung') 
ax1.loglog(n_p_array, tau_c, color = 'orange',linestyle = '-.', label = 'Coulomb scattering')

ax1.set_title('$E_e$ = 0.01 GeV', fontsize = 18)
ax1.set_ylabel(r'$\tau$ / Myr', fontsize = 18)
ax1.set_xlabel('$n_H$ / cm$^{-3}$', fontsize = 18)
ax1.tick_params(axis='both', which='major', labelsize = 16)

# plotting thermalization efficiency
ax2 = ax1.twinx() # create another y-axis sharing a common x-axis
ax2.plot(n_p_array, f_c, 'r', linestyle = '-')
ax2.set_ylabel('$f_C$', color = 'red', fontsize = 18)
ax2.set_ylim(-0.05, 1.05)
ax2.tick_params(axis = 'y', labelcolor = 'red', labelsize = 16)

#fig.subplots_adjust(top = 0.8)
#fig.legend()
plt.savefig('(0.01 GeV) tau vs. n_H.pdf')
```
